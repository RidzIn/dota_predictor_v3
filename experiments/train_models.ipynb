```python
from utils import features_dataset_winrates
from autogluon.tabular import TabularPredictor
import pandas as pd
import json

minor = pd.read_pickle('C:\\Users\\Ridz\\Desktop\\Matches\\minor_matches.pkl')
ti13 = pd.read_pickle('C:\\Users\\Ridz\\Desktop\\Matches\\ti13_matches.pkl')


with open('data/winrates_minor_v2.json') as file:
    winrates_dict = json.load(file)


features_minor = features_dataset_winrates(minor, winrates_dict)
features_ti13 = features_dataset_winrates(ti13, winrates_dict)

label_column = 30

predictor = TabularPredictor(label=label_column, eval_metric='accuracy').fit(
    train_data=features_minor,
    time_limit=3600,
    presets='best_quality',
    hyperparameters={
        'GBM': {},      # LightGBM
        'CAT': {},      # CatBoost
        'RF': {},       # Случайные леса
        'XT': {},       # Экстремальные деревья
        
    },
    hyperparameter_tune_kwargs={'num_trials': 10, 'scheduler': 'local', 'searcher': 'random'}
)

predictor.leaderboard(features_ti13)
```
